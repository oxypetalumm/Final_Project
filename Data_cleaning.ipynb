```python
import pandas as pd
import numpy as np
```

```python
# Read Excel Files
ca_2017_df = pd.DataFrame(pd.read_excel("Data/california_2017.xls"))
ca_2018_df = pd.DataFrame(pd.read_excel("Data/california_2018.xls"))
ca_2019_df = pd.DataFrame(pd.read_excel("Data/california_2019.xls"))

# view dataframes
display(ca_2017_df)
display(ca_2018_df)
display(ca_2019_df)
```

## Clean ca_2017_df

```python
# clean rows
cleaned_ca_2017_df_1 = ca_2017_df.drop(labels=[0, 1, 2, 3, 460, 461], axis=0)

# rename column headers
renamed_ca_2017_df = cleaned_ca_2017_df_1.rename(columns = {'Table 8': 'City', 
                                                            'Unnamed: 1': 'Population', 
                                                            'Unnamed: 2': 'Violent crime', 
                                                            "Unnamed: 3": "Murder",
                                                            "Unnamed: 4": "Rape", 
                                                            'Unnamed: 5': 'Robbery',
                                                            'Unnamed: 6': 'Aggravated assault',
                                                            'Unnamed: 7': 'Property crime', 
                                                            'Unnamed: 8': 'Burglary',
                                                            'Unnamed: 9': 'Larceny-theft', 
                                                            'Unnamed: 10': 'Motor vehicle theft',
                                                            'Unnamed: 11': 'Arson'})

# reset index
cleaned_ca_2017_df = renamed_ca_2017_df.reset_index(drop=True)

# print dataframe
cleaned_ca_2017_df
```
